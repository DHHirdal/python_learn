```python
import matplotlib.pyplot as plt
import numpy as np

xpoints =np.array([3,2,4,6,10])
ypoints =np.array([3,8,4,5,9])
plt.plot(ypoints, 'P:r', ms =20)
plt.plot(xpoints, 'o--g',ms=10)
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints =np.array([3,2,4,6,10])
ypoints =np.array([3,8,4,5,9])
plt.plot(ypoints, 'P:r', ms =20, markeredgecolor ='m', mfc='b')
plt.plot(xpoints, 'o--g',ms=10, markeredgecolor ='b', mfc ='c')
```
