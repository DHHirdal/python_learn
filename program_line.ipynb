Set Font Properties for Title and Labels
You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.


```python
import matplotlib as mt
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,2,3,10,5,1])
ypoints = np.array([4,8,2,8,0,5])

plt.plot(xpoints, ls='dashed', color ='g')
plt.plot(ypoints, ls='-.', color ='m')
fontdict1 ={'family':'serif', 'color':'r','size':15 }
fontdict2 ={'family':'serif', 'color':'b','size':20 }
plt.title("time-Distance graph", loc='right')
plt.xlabel ("distance", fontdict=fontdict1)
plt.ylabel("time",fontdict=fontdict2)
```
