```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(['a', 'b', 'c', 'd'])
y= np.array ([10,20,30,40])

plt.bar(x,y)
plt.show()

x=np.array(['a', 'b', 'c', 'd'])
y= np.array ([10,20,30,40])

plt.barh(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(['a', 'b', 'c', 'd'])
y= np.array ([10,20,30,40])

plt.bar(x,y, color ='g', width = 0.1) # color and width added
plt.show()

x=np.array(['a', 'b', 'c', 'd'])
y= np.array ([10,20,30,40])

plt.barh(x,y, color ='m', height=0.2) # color and Height is added to the function
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(['10-04-2025', '11-04-2025', '12-04-2025', '13-04-2025'])
y= np.array ([10,20,30,40])

plt.title("JIRA ticket  graph")
plt.xlabel("Date")
plt.ylabel("ticket count")
plt.bar(x,y, color ='g', width = 0.1) # color and width added
plt.show()

x=np.array(['10th-May-25', '11th-May-25', '12th-May-25', '13th-May-25'])
y= np.array ([10,20,30,40])

plt.title("JIRA ticket  graph")
plt.ylabel("Date", color = 'g')
plt.xlabel("ticket count", color = 'g')
plt.barh(x,y, color ='m', height=0.2) # color and Height is added to the function
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(['10-04-2025', '11-04-2025', '12-04-2025', '13-04-2025'])
y= np.array ([10,20,30,40])
colors = ['r', 'g', 'b', 'y']

plt.title("JIRA ticket  graph")
plt.xlabel("Date")
plt.ylabel("ticket count")
plt.bar(x,y, color =colors, width = 0.1) # each bar is different color
plt.legend(colors)
plt.show()

x=np.array(['10th-May-25', '11th-May-25', '12th-May-25', '13th-May-25'])
y= np.array ([10,20,30,40])
colors = ['r', 'g', 'b', 'y']

plt.title("JIRA ticket  graph")
plt.ylabel("Date", color = 'g')
plt.xlabel("ticket count", color = 'g')
plt.barh(x,y, color =colors, height=0.2) # each bar is different color
plt.legend()
plt.show()
```
